```python
import pandas as pd
import json

# Lire le fichier Excel
df = pd.read_excel("/content/last_final_list_en_msa_dar.xlsx")
df = df.dropna()

# Initialiser une liste pour stocker les traductions au format JSON
translations_json = []

# Itérer sur chaque ligne du DataFrame
for idx, row in df.iterrows():
    # Construire le dictionnaire de traduction pour cette ligne
    translation_dict = {
        "translation": {
            "idx": idx,
            "English": row["English Sentences"],
            "Darija": row["Darija Sentences"],
            "MSA": row["Modern Standard Arabic Sentences"]
        }
    }
    # Ajouter la traduction au format JSON à la liste
    translations_json.append(json.dumps(translation_dict, ensure_ascii=False))

# Écrire la liste des traductions au format JSON dans un fichier
with open("last_final_list_en_msa_dar2.json", "w", encoding="utf-8") as f:
    f.write("\n".join(translations_json))

print("Le fichier JSON a été créé avec succès !")
```

```python
import pandas as pd

# Lecture du fichier texte ligne par ligne
with open("C:/Users/pc/Downloads/test-4000.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Initialisation des listes pour chaque colonne
english_sentences = []
msa_sentences = []
darija_sentences = []

# Extraction des données de chaque ligne et ajout dans les listes correspondantes
for line in lines:
    data = eval(line)  # Convertir la ligne JSON en un dictionnaire Python
    english_sentences.append(data["translation"]["English"])
    msa_sentences.append(data["translation"]["MSA"])
    darija_sentences.append(data["translation"]["Darija"])

# Création du DataFrame pandas
df = pd.DataFrame({
    "English": english_sentences,
    "MSA": msa_sentences,
    "Darija": darija_sentences
})

# Enregistrer le DataFrame dans un fichier Excel
df.to_excel("C:/Users/pc/Downloads/test-4000.xlsx", index=False)
```
