```python
final_list = []  # Création de la liste finale vide
```

```python
def extract_python_list(text):
    start_index = text.find("[")
    end_index = text.find("]")

    if start_index == -1 or end_index == -1:
        return []  # Aucune liste trouvée, retourne une liste vide

    python_list = text[start_index:end_index+1]  # Extraction de la liste Python
    try:
        extracted_list = eval(python_list)  # Convertit la chaîne en liste Python
        if isinstance(extracted_list, list):
            return extracted_list
        else:
            return []  # Si le contenu entre crochets n'est pas une liste valide, retourne une liste vide
    except SyntaxError:
        return []  # En cas d'erreur de syntaxe, retourne une liste vide


def add_to_final_list(input_list):
    for j in range (len(input_list)) :
        if input_list[j] not in final_list:
            final_list.append(input_list[j])
    return
```

```python
import pandas as pd

def add_to_final_list(input_list, final_liste):
    for j in range (len(input_list)) :
        if input_list[j] not in final_liste:
            final_liste.append(input_list[j])
    return final_liste

def add_phrases_to_excel(file_path, phrases_list, output_file_path):
    # Charger le fichier Excel en une DataFrame
    df = pd.read_excel(file_path)

    # Convertir la colonne en une liste
    original_data = df['English Sentences'].tolist()

    # Ajouter les phrases de la liste à la liste originale
    updated_data = add_to_final_list(phrases_list, original_data)

    # Créer une nouvelle DataFrame avec les données mises à jour
    updated_df = pd.DataFrame({'English Sentences': updated_data})

    # Écrire la nouvelle DataFrame dans un fichier Excel
    updated_df.to_excel(output_file_path, index=False)
```

```python
# Utilisation de la fonction
file_path = '/content/final_list.xlsx'
output_file_path = '/content/final_list_2.xlsx'
phrases_list = english_sentences
add_phrases_to_excel(file_path, phrases_list, output_file_path)
```

```python
#Find Phrases in The little Prince
import re
def txt_to_sentences(file_path):
    # Ouvrir le fichier en mode lecture
    with open(file_path, 'r', encoding='latin-1') as file:
        # Lire tout le contenu du fichier
        content = file.read()

        # Diviser le contenu en phrases en utilisant ".", "?", et "!" comme délimiteurs
        sentences = [sentence.strip() for sentence in re.split(r'[.!?]', content) if sentence.strip()]

        return sentences

file_path = '/content/THE_LITTLE_PRINCE.txt'
sentences = txt_to_sentences(file_path)
print(len(sentences))

```

```python
# #Find Phrases in The Arabian Night

file_path = '/content/arabn10.txt'
sentences_ = txt_to_sentences(file_path)
print(len(sentences_))
```

```python
import re

def remove_chapter_numbers(sentences):
    # Expression régulière pour matcher "Chapter" suivi d'un espace et d'un nombre
    chapter_pattern = re.compile(r'Chapter\s+\d+', re.IGNORECASE)

    # Parcourir chaque phrase dans la liste
    cleaned_sentences = []
    for sentence in sentences:
        # Supprimer "Chapter x" de la phrase
        cleaned_sentence = re.sub(chapter_pattern, '', sentence)
        # Ajouter la phrase nettoyée à la nouvelle liste si elle n'est pas vide
        if cleaned_sentence.strip():
            cleaned_sentences.append(cleaned_sentence.strip())

    return cleaned_sentences
print("1___________________________________________________")
cleaned_sentences = remove_chapter_numbers(sentences)
print(print(cleaned_sentences))


print("2___________________________________________________")
cleaned_sentences_ = remove_chapter_numbers(sentences_)
print(print(cleaned_sentences_))
```

```python
def remove_newlines(sentences):
    cleaned_sentences = []
    for sentence in sentences:
        # Supprimer toutes les occurrences de '\n' dans la phrase
        cleaned_sentence = sentence.replace('\n', '')
        # Ajouter la phrase nettoyée à la nouvelle liste
        cleaned_sentences.append(cleaned_sentence)

    return cleaned_sentences

print("1___________________________________________________")
cleaned_sentences_2 = remove_newlines(cleaned_sentences)
print(cleaned_sentences_2)

print("1___________________________________________________")
cleaned_sentences_2_ = remove_newlines(cleaned_sentences_)
print(cleaned_sentences_2_)
```

```python
def find_longest_sentences(sentences):
    # Trier la liste des phrases par longueur décroissante
    sorted_sentences = sorted(sentences, key=len, reverse=True)
    # Récupérer les deux premières phrases de la liste triée
    longest_sentences = sorted_sentences[:2]
    return longest_sentences

# Exemple d'utilisation
longest_sentences = find_longest_sentences(cleaned_sentences_2)
print(longest_sentences)

longest_sentences_ = find_longest_sentences(cleaned_sentences_2_)
print(longest_sentences_)
```
