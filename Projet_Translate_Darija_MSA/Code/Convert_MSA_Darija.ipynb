```python
from transformers import SeamlessM4Tv2Model, AutoProcessor
from transformers import AutoProcessor

import torch
device = "cuda:0" if torch.cuda.is_available() else "cpu"

processor = AutoProcessor.from_pretrained("facebook/seamless-m4t-v2-large")
model = SeamlessM4Tv2Model.from_pretrained("facebook/seamless-m4t-v2-large")
model = model.to(device)
```

```python
import pandas as pd

def translate_to_arabic(input_file_path, output_file_path):
    # Charger le fichier Excel
    df = pd.read_excel(input_file_path)

    # Initialiser une liste pour stocker les traductions
    translations = []

    # Processus de traduction pour chaque phrase en anglais
    for sentence in df['English Sentences']:
        # Traitement de l'entrée
        text_inputs = processor(text=sentence, return_tensors="pt", src_lang="eng").to(device)

        # Générer la traduction
        output_tokens = model.generate(**text_inputs, tgt_lang="arb", generate_speech=False)
        translated_text_from_text = processor.decode(output_tokens[0].tolist()[0], skip_special_tokens=True)

        # Ajouter la phrase originale et sa traduction à la liste des traductions
        translations.append(translated_text_from_text)

    # Ajouter la liste des traductions au DataFrame
    df['Modern Standard Arabic Sentences'] = translations

    # Sauvegarder le DataFrame avec les traductions dans un fichier Excel
    df.to_excel(output_file_path, index=False)

    print("Translations saved to", output_file_path)
```

# Using the model on our data for MSA translation

```python
import pandas as pd

def translate_to_arabic(input_file_path, output_file_path_prefix, chunk_size=500):
    # Charger le fichier Excel
    df = pd.read_excel(input_file_path)

    # Diviser le DataFrame en morceaux de taille spécifiée
    chunks = [df[i:i+chunk_size] for i in range(0, len(df), chunk_size)]

    # Processus de traduction pour chaque morceau
    for i, chunk in enumerate(chunks):
        # Initialiser une liste pour stocker les traductions de ce morceau
        translations = []

        # Processus de traduction pour chaque phrase en anglais dans le morceau
        for sentence in chunk['English Sentences']:
            # Traitement de l'entrée
            text_inputs = processor(text=sentence, return_tensors="pt", src_lang="eng").to(device)

            # Générer la traduction
            output_tokens = model.generate(**text_inputs, tgt_lang="arb", generate_speech=False)
            translated_text_from_text = processor.decode(output_tokens[0].tolist()[0], skip_special_tokens=True)

            # Ajouter la traduction à la liste des traductions
            translations.append(translated_text_from_text)

        # Ajouter les traductions au DataFrame
        chunk['Modern Standard Arabic Sentences'] = translations

        # Sauvegarder le morceau avec les traductions dans un fichier Excel partiel
        output_file_path = f"{output_file_path_prefix}_{i+4}.xlsx"
        chunk.to_excel(output_file_path, index=False)

        print(f"Translations saved to {output_file_path}")
```
